# POS tagging

```python
#POS tagging for text #1
import spacy
nlp=spacy.load('en_core_web_sm')
text1 = 'The average person squanders about forty days a year compensating for things he or she has forgotten'
for token in nlp(text1):
    print(token.text, '--->',token.pos_,'--->',token.tag_)
```

```python
#POS tagging for text #2
import spacy
nlp=spacy.load('en_core_web_sm')
text2 = 'It was simply a matter of learning to “think in more memorable ways” using the “extraordinarily simple” 2,500-year-old mnemonic technique known as the “memory palace” that Simonides of Ceos had supposedly invented in the rubble of the great banquet hall collapse.'
for token in nlp(text2):
    print(token.text, '--->',token.pos_,'--->',token.tag_)
```

# Dependency Parsing

```python
#Dependency Parsing for text #1
from spacy import displacy
text1 = 'The average person squanders about forty days a year compensating for things he or she has forgotten'
for token in nlp(text1):
    print(token.text, '--->',token.dep_,'--->',token.head.text)
displacy.render(nlp(text1),jupyter=True)
```
